```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
df['Family Size']=df['SibSp']+df['Parch']+1
df.head()
```

```python
dff=df.drop(['PassengerId','Name','Ticket','Cabin','Survived','SibSp','Parch'],axis='columns')
dff.head()
```

```python
target=df['Survived']
target.head()
```

```python
dff['Embarked'] = dff['Embarked'].fillna(dff['Embarked'].mode()[0])
dff['Fare'] = dff['Fare'].fillna(dff['Fare'].mean())

dff.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex=LabelEncoder()
le_Embarked=LabelEncoder()
```

```python
inputs=dff
inputs['Sex_n']=le_Sex.fit_transform(inputs['Sex'])
inputs['Embarked_n']=le_Embarked.fit_transform(inputs['Embarked'])
inputs.head()
```

```python
final=dff.drop(['Sex','Embarked'],axis='columns')
final.head()
```

```python
age_known=final[final.Age.notnull()]
age_unknown=final[final.Age.isnull()]
```

```python
features=['Pclass','Fare','Sex_n','Embarked_n','Family Size']
x_known=age_known[features]
y_known=age_known['Age']
x_unknown=age_unknown[features]
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_known,y_known)
```

```python
predicted_age=model.predict(x_unknown)
predicted_age
```

```python
print(final['Age'].isnull().sum())
print(len(predicted_age))
```

```python
final.loc[final['Age'].isnull(),'Age']=predicted_age
final.head()
```

```python
print(final['Age'].isnull().sum())
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(final,target,test_size=0.2)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
